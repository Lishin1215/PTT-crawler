# Python crawler for PTT studyabroad 留學版

```python
import time
import requests
import bs4
import xlwings as xw
search_tab_number = 20
start = 1725
end = start - search_tab_number
keyword_list = []

#=======================================================================================
# 讓使用者輸入關鍵字，輸入end代表結束
#=======================================================================================
while 1:
    keyword = str(input('請輸入關鍵字'))
    if keyword == 'end':
        break
    else:
        keyword_list.append(keyword)
#=======================================================================================
#=======================================================================================

#=======================================================================================
# 新增一個excel worksheet
#=======================================================================================
wb = xw.Book()
sheet = wb.sheets["工作表1"]
sheet.range("A1").value = "標題"
sheet.range("B1").value = "網址"
row = 2
#=======================================================================================
# 在各個分頁中找到標題，並看標題裡面有沒有包含keyword
#=======================================================================================
for page in range(start, end, -1):
    current_tab_url = 'https://www.ptt.cc/bbs/studyabroad/index' + str(page) + '.html'
    res = requests.get(current_tab_url)
    html = bs4.BeautifulSoup(res.text, "html.parser")
    title_list = html.findAll('div', 'title')
    # 如果循環跑太快，PTT網站可能會受不了，因此用time.sleep(1)讓整個程式暫停1秒
    time.sleep(1)
    for title in title_list:
        try:
            # 因為有很多關鍵字要檢查，因此要用for迴圈去一一取出keyword_list中的keyword
            for keyword in keyword_list:
                # 檢查這個迴圈中的keyword是否再title.a.string中
                if keyword in title.a.string:
                    print(title.a.string)
                    print("https://www.ptt.cc"+title.a.get("href"))#取得文章連結
                    
                    sheet.range(f"A{row}").value = title.a.string
                    sheet.range(f"B{row}").value = "https://www.ptt.cc"+title.a.get("href")
                    row = row + 1
                    
                    break
        except:
            continue
```
